```python
#Решение

documents = [
        {"type": "passport", "number": "2207 876234", "name": "Василий Гупкин"},
        {"type": "invoice", "number": "11-2", "name": "Геннадий Покемонов"},
        {"type": "insurance", "number": "10006", "name": "Аристарх Павлов"}
      ]
 
directories = {
        '1': ['2207 876234', '11-2'],
        '2': ['10006'],
        '3': []
      }

def get_name(docs, number1):
    str='Документ не найден в базе'
    for data in docs:
        if data["number"] == number1:
          str='Владелец документа: '+ data['name'] 
          break
    return str
 
def get_shelf(dirs, number): 
    for key, value in dirs.items():
        if number in value:
            return 'полка хранения: '+ key
    return 'В полках документа с данным номером нет.'
 
 
def get_list(docs):
    for doc in docs:
        print('№: ', doc['number'], 'тип: ', doc['type'], 'владелец: ',  doc['name'], get_shelf(directories, doc['number']))
        #print(doc['type'], doc['number'], doc['name'], get_shelf(directories, doc['number']))
 
 
def add_shelf(dirs, number):
    dirs.setdefault(number,[])
    print('Полка добавлена. Текущий перечень полок: ',', '.join(dirs.keys()))



def print_list_dirs(dirs):
    return 'Текущий перечень полок: ',', '.join(dirs.keys())


def del_shelf(dirs, delete_number):
#проверяем что такая полка есть и в ней нет документов
    i=0 
    if delete_number in dirs:
        for key, value in dirs.items(): 
            if delete_number == key and len(value)==0:
                i+=1
                del dirs[delete_number]
                print('Полка удалена. Текущий перечень полок: ',', '.join(dirs.keys()))
                break
    
        if i==0:
            print('на полке есть документа, удалите их перед удалением полки. Текущий перечень полок: ',', '.join(dirs.keys()))

    else:
        print('Такой полки не cуществует.Текущий перечень полок: ',', '.join(dirs.keys()))
        
    
 
def main (docs=documents,dirs=directories):
    while True:
        print('Возможные команды: p, s, l, as', 'ds', 'q')
        comand = input('Введите название команды ')
 
        if comand == 'p':
            num = input('Введите номер документа ')
            print(get_name(docs, num))
    
        elif comand == 's':
            num = input('Введите номер документа ')
            print(get_shelf(dirs, num))
    
        elif comand == 'l':
            get_list(docs)
        
        elif comand=='ds':
            shelf = input('Введите номер полки')
            del_shelf(dirs,shelf)

        elif comand == 'as':
            shelf = input('Введите номер полки.')
            if shelf not in dirs.keys():
                add_shelf(dirs, shelf)
            else:
                print('Такая полка уже существует.  Текущий перечень полок: ',', '.join(directories.keys()))
        
        elif comand == 'q':
            print('До свидания')
            break                
        
        else: print ('Документ не найден в базе')                
    return                     
  

main()
```
